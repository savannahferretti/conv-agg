## Plot Fig.1

For each case, open the corresponding .grd file, calculate distance between levels $\Delta z$ , and plot $\Delta z$ with height.

```python
cmap = np.asarray([[211,62,67,255],[248,196,98,255],[152,193,217,255],[64,72,119,255]])/255

fig,ax = pplt.subplots(width=2.72,height=2.6)
ax.format(xlabel='$\Delta$z (m)',xlim=(0,1600),ylabel='z (km)',ylim=(0,29))

for i,nz in enumerate(nzlist):
    # Open grid file
    filename = path+filename+str(nz)
    with open(filename) as file:
        grd = file.readlines()
        grd = [float(line.rstrip()) for line in grd]
    # Calculate and plot dz
    grd = np.asarray(grd)
    dz  = np.diff(grd)
    ax.plot(dz,grd[1:]/1000,color=cmap[i],label='nz_'+str(nz))

ax.legend(loc='r',ncols=1)
```
